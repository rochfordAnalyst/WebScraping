```python
until_list = ['2021-11-06','2021-11-09', '2021-11-10', '2021-11-12']
since_list = ['2021-11-04','2021-11-07', '2021-11-08', '2021-11-11']
```

```python
hashtag2 = 'boredape'

until2 = ' until:'
since2 = ' since:'
lang2 = ' lang:en'
bored_ape_tweets = pd.DataFrame()

for Udate, Sdate in zip(until_list, since_list):
    Search2 = hashtag2 + until2 + Udate + since2 + Sdate + lang2
    tweets_zz = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(Search2).get_items(), 100))
    bored_ape_tweets = bored_ape_tweets.append(tweets_zz)
#bored_ape_tweets
```

```python
tweetsWithSent_bored_ape = []
for t in bored_ape_tweets['content']:
    text = (t)
    ps = analyzer.polarity_scores(text)
    tweetsWithSent_bored_ape.append(ps['compound'])

bored_ape_tweets['compound score'] = tweetsWithSent_bored_ape
#bored_ape_tweets.describe()
```

```python
bored_ape_tweets['compound score'] = tweetsWithSent_bored_ape
#bored_ape_tweets
```

```python
df_master = pd.read_excel('nft master with sentiment.xlsx')
df_master
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy import stats
import seaborn as sns
from sklearn import linear_model
```

```python
y3 = np.array(df_master.thirty_day_change).reshape((-1, 1))
x3 = df_master[['Avg Compound score']]
```

```python
model_3 = LinearRegression().fit(x3, y3)
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
r_sq3 = model_3.score(x3, y3)
print('coefficient of determination:', r_sq3)

# Print the Intercept:
print('intercept:', model_3.intercept_)

# Print the Slope:
print('slope:', model_3.coef_) 

# Predict a Response and print it:
y_pred3 = model_3.predict(x3)
print('Predicted response:', y_pred3, sep='\n')
```
